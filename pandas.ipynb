# Read files using Pandas

```python
import pandas as pd
```

```python
#Reading file .CSV

pokemon = pd.read_csv('pokemon_data.csv')
pokemon.head()
```

```python
#Read headers
pokemon.columns

#Show especific column with the fist 5 position

print(pokemon['Name'][0:5])
```

```python
#Find especific object

poison = pokemon.loc[pokemon['Type 2'] == "Poison"]
poison
```

```python
#Show size

poison.size
```

```python
# Show describe

poison.describe()
```

```python
#Sorting values in alphabetical name

poison.sort_values(['Name'])
```

# Making changes

```python
#Adding total

pokemon['Total_Power'] = pokemon['HP'] + pokemon['Attack'] + pokemon['Sp. Atk'] + pokemon['Sp. Def'] + pokemon['Speed']
pokemon.head(5)
```

```python
#Drop columns

pokemon.drop(columns=['Legendary'])
```

# Filtering Data

```python
data_filtering = pokemon.loc[(pokemon['Type 1'] == 'Rock') & (pokemon['Attack'] > 100)]

data_filtering
```

# Making new file

```python
data_filtering.to_csv('data_filtering.csv')
```
